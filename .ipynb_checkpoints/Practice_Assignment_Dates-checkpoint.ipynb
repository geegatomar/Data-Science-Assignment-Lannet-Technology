```python
date_columns = []

for (i, column) in enumerate(df.columns):
    #print(type(df[column]))
    #print(df[column])
    #print(list(df[column]))
    values_in_column = list(df[column])
    for ele in values_in_column:
        #print(ele)
        ele = ele.replace(' ', '-')
        ele = ele.replace('/', '-')
        #print(ele)
        if ele.count('-') == 2 and i not in date_columns:
            date_columns.append(i)
                     
    #print(values_in_column)
print(date_columns)
```

```python
date_columns = set()

for (i, column) in enumerate(df.columns):
    
    values_in_column = list(df[column])
    
    for ele in values_in_column:
        
        try:
            ele = parse(ele)
            date_columns.add(i)
        except:
            pass
                     
    #print(values_in_column)
date_columns = list(date_columns)
print(date_columns)
```

```python
# We have successfully obtained all columns with dates in them
```

```python
for date_column in date_columns:
    print(date_column)
    #print(df[date_column])
    #print(df[df.columns[date_column]])
    dates_list = list(df[df.columns[date_column]])
    for (i, each_date) in enumerate(dates_list):
        dates_list[i] = parse(each_date).date()
        #print(each_date)
    print(dates_list)     
    df[df.columns[date_column]] = dates_list
```

```python
df = pd.read_csv('dates_problem.csv')
```

```python
df
```

```python
from dateutil.parser import *

date_columns = set()

for (i, column) in enumerate(df.columns):
    
    values_in_column = list(df[column])
    
    for ele in values_in_column:
        
        try:
            ele = parse(ele)
            date_columns.add(i)
        except:
            pass
                     
    #print(values_in_column)
date_columns = list(date_columns)
print(date_columns)
```

```python
from datetime import date

for date_column in date_columns:
    print(date_column)
    #print(df[date_column])
    #print(df[df.columns[date_column]])
    dates_list = list(df[df.columns[date_column]])
    for (i, each_date) in enumerate(dates_list):
        try:
            dates_list[i] = parse(each_date).date()
            # parse all the 'str' dates into datetime dates format
        except:
            dates_list[i] = date.today()
            # if it throws an exception, we will assign it todays date. It can be changed according to the quesion
            
        #print(each_date)
    print(dates_list)     
    df[df.columns[date_column]] = dates_list
```
