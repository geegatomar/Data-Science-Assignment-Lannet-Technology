```python
import pandas as pd

df = pd.read_csv('heart.csv')
```

```python
from scipy.stats import pearsonr

interested = []

for first_col in df.columns:
    for second_col in df.columns:
        if(first_col != second_col):
            corr, p_value = pearsonr(df[first_col], df[second_col])
            if corr > 0.35:
                interested.append(first_col)
                interested.append(second_col)
                print(corr, first_col, second_col)
```

```python
from scipy.stats import pearsonr

interested = []
no_of_corr = dict()
its_correlators = dict()


for first_col in df.columns:
    for second_col in df.columns:
        if(first_col != second_col):
            corr, p_value = pearsonr(df[first_col], df[second_col])
            
            if corr > 0.85:
                interested.append(first_col)
                print(corr, first_col, second_col)
                if first_col not in no_of_corr:
                    no_of_corr[first_col] = 1
                else:
                    no_of_corr[first_col] += 1
               
                if first_col in its_correlators:
                    its_correlators[first_col].append(second_col)
                else:
                    its_correlators[first_col] = [second_col]
                    
               
                    
print(no_of_corr)
print(its_correlators)
```

```python
columns_to_drop = []

while len(no_of_corr) > 0:
    max_corr = max(no_of_corr, key = no_of_corr.get)
    print(max_corr)
    columns_to_drop.append(max_corr)
    for each_correlator in its_correlators[max_corr]:
        if each_correlator in no_of_corr:
            no_of_corr[each_correlator] -= 1
    print(no_of_corr)
    print(its_correlators)
    for key in list(no_of_corr):
        if no_of_corr[key] <= 0:
            no_of_corr.pop(key, 'None')
            
   # for (k, v) in no_of_corr:
    #    print(k ,v)
     #   if v <= 0:
      #      print(k, v)
       #     no_of_corr.pop(k, 'None')
    no_of_corr.pop(max_corr, 'None')
    print(no_of_corr)
    print(its_correlators)
        
```

```python
print(columns_to_drop)
```

```python
df.drop(columns_to_drop, axis="columns")
```

```python
dataset = df
threshold = 0.85


def correlation(dataset, threshold):
    col_corr = set() # Set of all the names of deleted columns
    corr_matrix = dataset.corr()
    print(col_corr)
    print(corr_matrix)
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j] >= threshold) and (corr_matrix.columns[j] not in col_corr):
                colname = corr_matrix.columns[i] # getting the name of column
                col_corr.add(colname)
                if colname in dataset.columns:
                    del dataset[colname] # deleting the column from the dataset

    print(dataset)
    return dataset
```
