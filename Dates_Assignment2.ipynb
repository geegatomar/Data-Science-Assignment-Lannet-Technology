```python
import pandas as pd

df = pd.read_csv('dates.csv')
```

```python
df
```

```python
from dateutil.parser import *

date_columns = []

for (i, column) in enumerate(df.columns):
    
    values_in_column = list(df[column])
    
    for ele in values_in_column:
        
        try:
            ele = parse(ele)
            date_columns.append(i)
            break
            
        except:
            pass
                     
            
print(date_columns)
```

```python
# We have successfully obtained all columns having dates in them
```

```python
from datetime import date


for date_column in date_columns:
    #print(date_column)
    dates_list = list(df[df.columns[date_column]])
    for (i, each_date) in enumerate(dates_list):
        try:
            dates_list[i] = parse(each_date).date()
        except:
            dates_list[i] = date.today()
            
        #print(each_date)
    print(dates_list)     
    df[df.columns[date_column]] = dates_list
```

```python
df
```

```python
import itertools

combinations = list(itertools.combinations(date_columns, 2))
combinations
```

```python
for each_combi in combinations:
    col1, col2 = each_combi
    col1_name = df.columns[col1]
    col2_name = df.columns[col2]
    df['Diff_of_{}_and_{}'.format(col1_name, col2_name)] = df[df.columns[col1]] - df[df.columns[col2]]
    # now we have obtained difference with value of type datetime.timedelta
    # now converting it to int(number of days)
    
    df['Diff_of_{}_and_{}_in_days'.format(col1_name, col2_name)] =df['Diff_of_{}_and_{}'.format(col1_name, col2_name)].apply(lambda x : abs(x.days))
```
